```python
# 데이터 준비하기
from google.colab import files
ul = files.upload()
```

```python
import numpy as np
import io
import pandas as pd

data = np.array(pd.read_csv(io.BytesIO(ul['fin2dat.csv'])))
data = list(data)
data.append([0.436,0.0259])    ### 0번째 index가 마지막 index로 들어감
data = np.array(data)
print(data)
```

```python
import matplotlib.pyplot as plt

plt.scatter(data[:,0],data[:,1])
plt.show()
```

# (b) 
**K-Means 클러스터링 방법을 이용하여 $K=2$의 두 그룹으로 분류하고, 서로 다른 클러스터에 속한 점은 다른 색으로 표시하여 그래프로 나타내어라.**

```python
# K-means 클러스터링
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=2)
result = kmeans.fit_predict(data)
print(result)

# 두 그룹으로 분류
group1 = [] ; group2 = []
for i in range(len(result)):
  if result[i] == 0:
    group1.append(data[i])
  else:
    group2.append(data[i])

group1 = np.array(group1) ; group2 = np.array(group2)

# 그래프 그리기
import matplotlib.pyplot as plt
plt.scatter(group1[:,0],group1[:,1],c='r')
plt.scatter(group2[:,0],group2[:,1],c='b')
plt.legend(['Group 1', 'Group 2'])
plt.show()

# 각 cluster의 중심점 찾기
center1 = [np.mean(group1[:,0]), np.mean(group1[:,1])]
center2 = [np.mean(group2[:,0]), np.mean(group2[:,1])]
print('Center Point of Cluster 1 =',center1)
print('Center Point of Cluster 2 =',center2)
```

# (c)
**$K = 3$과 $4$에 대하여 (b)를 반복하라. 또 위 데이터는 $K = 2,3,4$중 몇 개의 클러스터로 분류하는 것이 가장 적절한 지 프로그램의 뒷 부분에 text로 기술하라.**

```python
# K-means 클러스터링 (K=3)
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=3)
result = kmeans.fit_predict(data)
print(result)

# 세 그룹으로 분류
group1 = [] ; group2 = [] ; group3 = []
for i in range(len(result)):
  if result[i] == 0:
    group1.append(data[i])
  elif result[i] == 1:
    group2.append(data[i])
  else:
    group3.append(data[i])

group1 = np.array(group1) ; group2 = np.array(group2) ; group3 = np.array(group3)

# 그래프 그리기
import matplotlib.pyplot as plt
plt.scatter(group1[:,0],group1[:,1],c='r')
plt.scatter(group2[:,0],group2[:,1],c='g')
plt.scatter(group3[:,0],group3[:,1],c='b')
plt.legend(['Group 1', 'Group 2','Group 3'])
plt.show()
```

```python
# K-means 클러스터링 (K=4)
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=4)
result = kmeans.fit_predict(data)
print(result)

# 네 그룹으로 분류
group1 = [] ; group2 = [] ; group3 = [] ; group4 = []
for i in range(len(result)):
  if result[i] == 0:
    group1.append(data[i])
  elif result[i] == 1:
    group2.append(data[i])
  elif result[i] == 2:
    group3.append(data[i])
  else: group4.append(data[i])

group1 = np.array(group1) ; group2 = np.array(group2) ; group3 = np.array(group3) ; group4 = np.array(group4)

# 그래프 그리기
import matplotlib.pyplot as plt
plt.scatter(group1[:,0],group1[:,1],c='r')
plt.scatter(group2[:,0],group2[:,1],c='g')
plt.scatter(group3[:,0],group3[:,1],c='b')
plt.scatter(group4[:,0],group4[:,1],c='black')
plt.legend(['Group 1', 'Group 2','Group 3','Group 4'])
plt.show()
```

**K = 4** 가 적절한 것 같다. 
=> 위와 아래, 오른쪽과 왼쪽이 점들이 각자 나눠져있는 것처럼 봉기 때문이다.

# (d)
**$K = 2$인 경우. (b)에서 나온 결과 다른 최종 클러스터가 되는 초기 cluster center을 찾고, 이를 이용한 최종 분류 결과를 그래프로 나타내어라.**

```python
# K-means 클러스터링
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=2)
result = kmeans.fit_predict(data)
print(result)

# 두 그룹으로 분류
group1 = [] ; group2 = []
for i in range(len(result)):
  if result[i] == 0:
    group1.append(data[i])
  else:
    group2.append(data[i])

group1 = np.array(group1) ; group2 = np.array(group2)

# 그래프 그리기
import matplotlib.pyplot as plt
plt.scatter(group1[:,0],group1[:,1],c='r')
plt.scatter(group2[:,0],group2[:,1],c='b')
plt.legend(['Group 1', 'Group 2'])
plt.show()

# 각 cluster의 중심점 찾기
center1 = [np.mean(group1[:,0]), np.mean(group1[:,1])]
center2 = [np.mean(group2[:,0]), np.mean(group2[:,1])]
print('Center Point of Cluster 1 =',center1)
print('Center Point of Cluster 2 =',center2)
```

# (e)
**위 데이터에 대한 주성분 분석(PCA)를 실시하려고 한다. 1차원 주성분 벡터(단위 벡터의 $x, y$ 성분)을 구하고, 40개 데이터를 주성분 벡터로 투사 시킨 값을 구하라.**

```python
M = np.mean(data, axis=0)   ### 평균
C = data-M   ### 편차
V = np.cov(C.T)   ### 공분산 행렬
eva, eve = np.linalg.eig(V)   ### 고유값, 고유벡터

for i in range(2):
  print(eva[i], ":", eve.T[i])
  
print('\n')

P = eve.T.dot(data.T) ; print(P)
```
