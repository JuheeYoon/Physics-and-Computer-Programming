# (b)
**개인별, 학기별 평균 평점을 출력하라. (단, 과목 평점에 학점을 가중하여 학기별 평균을 구하고 Pass NonPass 과목은 제외하라.)**

```python
# 학생 번호가 시작하는 index
start_index = []
for i in range(len(data['NO'])):
  if str(data['NO'][i]) != 'nan':
    start_index.append(i)
print(start_index)

# 학생 넘버
student = []
for i in range(len(data['NO'])):
  if i in start_index:
    student.append(data['NO'][i])
print(student)
```

```python
# 학생별 학기가 시작하는 index
sem_index = [[] for i in range(len(start_index))]
n = -1
data['수강학기 '].astype(str)

for i in range(len(data['수강학기 '])):
  if i in start_index: n += 1
  if str(data['수강학기 '][i]) != 'nan':
    sem_index[n].append(i)
print(sem_index)
```

```python
# P/NP 과목이 있는 index
PNP_index = []
for i in range(len(data['등급'])):
  if (data['등급'][i] == 'P') or (data['등급'][i] == 'NP') : PNP_index.append(i)
print(PNP_index)
```

```python
Full = [[] for i in range(18)]   # 18명의 학생

for i in range(len(Full)):
  for j in range(len(sem_index[i])):
    Full[i].append([])

n = -1
for i in range(len(data['등급'])): #데이터 한줄씩 읽기
  if i in start_index: n+=1 ; m = -1 # 인덱스
  if i in sem_index[n]: m+=1 ; Full[n][m].append(data['수강학기 '][i]) ; Full[n][m].append(0) ; Full[n][m].append(0)
  if i not in PNP_index:
    Full[n][m][1] += data['학점'][i] * data['평점'][i]
    Full[n][m][2] += data['학점'][i]

for i in range(len(student)):
  print('\n')
  print('학생 NO : ',student[i])
  for j in range(len(Full[i])):
    if Full[i][j][2] == 0 : continue
    print('%s학기 -> 평균 평점 = %f' %(Full[i][j][0], (Full[i][j][1]/Full[i][j][2])))
    




  

```

```python
dic = {}
for i in range(len(data['수강과목'])):
  if data['수강과목'][i] == '물리천문입문1': data['수강과목'][i] = '물리학입문1'
  if data['수강과목'][i] == '물리천문입문2': data['수강과목'][i] = '물리학입문2'
  if data['수강과목'][i] in dic:
    dic[data['수강과목'][i]] += 1
  else:
    dic[data['수강과목'][i]] = 1

print(dic)
```
