***. 퍼텐셜내에서의 운동 : 뉴턴의 법칙
  $⇒$ 포물선 운동

```python
import numpy as np
import matplotlib.pyplot as plt

g=9.8
v0=10
theta=45*(np.pi/180) #라디안
dt=0.01

vx=v0*np.cos(theta)
vy=v0*np.sin(theta)

x=0
y=0
px=[]
py=[]
while y > -1e-10:
  x+=vx*dt
  y+=vy*dt
  vy-=g*dt
  px.append(x)
  py.append(y)

plt.plot(px,py)
plt.show()
```

***. 퍼텐셜내에서의 운동 : 뉴턴의 법칙 ( 동영상 )

```python
import numpy as np
import matplotlib.pyplot as plt

g=9.8
v0=10
theta=45*(np.pi/180) #라디안
dt=0.01

vx=v0*np.cos(theta)
vy=v0*np.sin(theta)

x=0
y=0
px=[]
py=[]
while y >= 0:
  x+=vx*dt
  y+=vy*dt
  vy-=g*dt

  px.append(x)
  py.append(y)

# 동영상 만들기
from matplotlib import rc
rc('animation',html='jshtml')
from matplotlib.animation import FuncAnimation

fig,ax = plt.subplots()
xdata=[]
ydata=[]
ln, = plt.plot([],[],'ro')
def init():
  ax.set_xlim(min(px),max(px))
  ax.set_ylim(min(py),max(py))
  return ln,

def update(frame):
  xdata.append(px[frame])
  ydata.append(py[frame])
  ln.set_data(xdata,ydata)
  return ln,

ani=FuncAnimation(fig,update,frames=range(0,len(px),5),init_func=init,blit=False)
ani
```

```python
import numpy as np
import matplotlib.pyplot as plt

gamma = 0.1 ; g = 9.8 ; h = 30000 ; v0 = 100
dt = 0.01 ; t = 0

vx = v0 ; vy = 0
x = 0 ; y = h
px = [] ; py = []
while y > -1e-10:
  x += vx*dt
  y += vy*dt
  vx -= gamma*vx*dt
  vy -= (gamma*vy+g)*dt
  px.append(x)
  py.append(y)
  t += dt
plt.plot(px,py)
plt.show()

print(t,px[-1])
```
