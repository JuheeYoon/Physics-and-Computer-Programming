```python
smv = {'district':sm03['중앙선거관리위원회 선거통계시스템'],
       'p1':sm01['Unnamed: 1'], 'p2':sm02['Unnamed: 1'], 'p3':sm03['Unnamed: 1'], 'p4':sm04['Unnamed: 1'], 'p5':sm05['Unnamed: 1'], 'p6':sm06['Unnamed: 1'], 'p7':sm07['Unnamed: 1'], 'p7.5':sm075['Unnamed: 1'], # population 선거인수
       't1':sm01['Unnamed: 2'], 't2':sm02['Unnamed: 2'], 't3':sm03['Unnamed: 2'], 't4':sm04['Unnamed: 2'], 't5':sm05['Unnamed: 2'], 't6':sm06['Unnamed: 2'], 't7':sm07['Unnamed: 2'], 't7.5':sm075['Unnamed: 2'], # turnout 투표수
       'c1':sm01['Unnamed: 3'], 'c2':sm02['Unnamed: 3'], 'c3':sm03['Unnamed: 3'], 'c4':sm04['Unnamed: 4'], 'c5':sm05['Unnamed: 3'], 'c6':sm06['Unnamed: 3'], 'c7':sm07['Unnamed: 4'], 'c7.5':sm075['Unnamed: 4'], # conservative 보수
       'l1':sm01['Unnamed: 4'], 'l2':sm02['Unnamed: 4'], 'l3':sm03['Unnamed: 4'], 'l4':sm04['Unnamed: 3'], 'l5':sm05['Unnamed: 4'], 'l6':sm06['Unnamed: 4'], 'l7':sm07['Unnamed: 3'], 'l7.5':sm075['Unnamed: 3']} # liberal 진보

df = (pd.DataFrame(smv)[5:-2:2]).reset_index(drop=True)
df
```

```python
for i in df.columns:
  if i == 'district': continue
  else:
    df[i] = df[i].str.replace(',','').astype(str)
    df[i] = df[i].astype(int)

# 투표율
df['rt1'] = df['t1'] / df['p1']
df['rt2'] = df['t2'] / df['p2']
df['rt3'] = df['t3'] / df['p3']
df['rt4'] = df['t4'] / df['p4']
df['rt5'] = df['t5'] / df['p5']
df['rt6'] = df['t6'] / df['p6']
df['rt7'] = df['t7'] / df['p7']
df['rt7.5'] = df['t7.5'] / df['p7.5']

# 전체 투표 중 보수 투표율
df['rc1/e'] = df['c1'] / df['t1']
df['rc2/e'] = df['c2'] / df['t2']
df['rc3/e'] = df['c3'] / df['t3']
df['rc4/e'] = df['c4'] / df['t4']
df['rc5/e'] = df['c5'] / df['t5']
df['rc6/e'] = df['c6'] / df['t6']
df['rc7/e'] = df['c7'] / df['t7']
df['rc7.5/e'] = df['c7.5'] / df['t7.5']

# 전체 투표 중 진보 투표율
df['rl1/e'] = df['l1'] / df['t1']
df['rl2/e'] = df['l2'] / df['t2']
df['rl3/e'] = df['l3'] / df['t3']
df['rl4/e'] = df['l4'] / df['t4']
df['rl5/e'] = df['l5'] / df['t5']
df['rl6/e'] = df['l6'] / df['t6']
df['rl7/e'] = df['l7'] / df['t7']
df['rl7.5/e'] = df['l7.5'] / df['t7.5']

# 전체 투표 중 나머지 투표율
df['rr1/e'] = (df['t1'] - df['l1'] - df['c1']) / df['t1']
df['rr2/e'] = (df['t2'] - df['l2'] - df['c2']) / df['t2']
df['rr3/e'] = (df['t3'] - df['l3'] - df['c3']) / df['t3']
df['rr4/e'] = (df['t4'] - df['l4'] - df['c4']) / df['t4']
df['rr5/e'] = (df['t5'] - df['l5'] - df['c5']) / df['t5']
df['rr6/e'] = (df['t6'] - df['l6'] - df['c6']) / df['t6']
df['rr7/e'] = (df['t7'] - df['l7'] - df['c7']) / df['t7']
df['rr7.5/e'] = (df['t7.5'] - df['l7.5'] - df['c7.5']) / df['t7.5']

# 진보+보수 중 보수 투표율
df['rc1'] = df['c1'] / ( df['c1'] + df['l1'] )
df['rc2'] = df['c2'] / ( df['c2'] + df['l2'] )
df['rc3'] = df['c3'] / ( df['c3'] + df['l3'] )
df['rc4'] = df['c4'] / ( df['c4'] + df['l4'] )
df['rc5'] = df['c5'] / ( df['c5'] + df['l5'] )
df['rc6'] = df['c6'] / ( df['c6'] + df['l6'] )
df['rc7'] = df['c7'] / ( df['c7'] + df['l7'] )
df['rc7.5'] = df['c7.5'] / ( df['c7.5'] + df['l7.5'] )

# 진보+보수 중 진보 투표율
df['rl1'] = df['l1'] / ( df['c1'] + df['l1'] )
df['rl2'] = df['l2'] / ( df['c2'] + df['l2'] )
df['rl3'] = df['l3'] / ( df['c3'] + df['l3'] )
df['rl4'] = df['l4'] / ( df['c4'] + df['l4'] )
df['rl5'] = df['l5'] / ( df['c5'] + df['l5'] )
df['rl6'] = df['l6'] / ( df['c6'] + df['l6'] )
df['rl7'] = df['l7'] / ( df['c7'] + df['l7'] )
df['rl7.5'] = df['l7.5'] / ( df['c7.5'] + df['l7.5'] )

# 투표율 평균
df['t_ave'] = df.loc[:,'rt1':'rt7.5'].mean(axis=1)
# 전체 투표 중 보수 투표율 평균
df['c/e_ave'] = df.loc[:,'rc1/e':'rc7.5/e'].mean(axis=1)
# 전체 투표 중 진보 투표율 평균
df['l/e_ave'] = df.loc[:,'rl1/e':'rl7.5/e'].mean(axis=1)
# # 전체 투표 중 나머지 투표율 평균
df['r/e_ave'] = df.loc[:,'rr1/e':'rr7.5/e'].mean(axis=1)
# 진보+보수 중 보수 투표율의 평균
df['c_ave'] = df.loc[:,'rc1':'rc7.5'].mean(axis=1)
# 진보+보수 중 진보 투표율의 평균
df['l_ave'] = df.loc[:,'rl1':'rl7.5'].mean(axis=1)

df
```

```python
# 구별 보수 투표율 dataframe
df2 = (sm.loc[1:,['district','rc1/e','rc2/e','rc3/e','rc4/e','rc5/e','rc6/e','rc7/e','rc7.5/e','c/e_ave']]).sort_values(by='c/e_ave')
df2 = df2.reset_index(drop=True)
# 구별 진보 투표율 dataframe
df3 = (sm.loc[1:,['district','rl1/e','rl2/e','rl3/e','rl4/e','rl5/e','rl6/e','rl7/e','rl7.5/e','l/e_ave']]).sort_values(by='l/e_ave')
df3 = df3.reset_index(drop=True)
# 구별 나머지 투표율 dataframe
df4 = (sm.loc[1:,['district','rr1/e','rr2/e','rr3/e','rr4/e','rr5/e','rr6/e','rr7/e','rr7.5/e','r/e_ave']]).sort_values(by='r/e_ave')
df4 = df4.reset_index(drop=True)

# 구별 구별 투표율 그래프
x1 = df2['district']
rcelist = ['rc1/e','rc2/e','rc3/e','rc4/e','rc5/e','rc6/e','rc7/e','rc7.5/e']
x2 = df3['district']
rlelist = ['rl1/e','rl2/e','rl3/e','rl4/e','rl5/e','rl6/e','rl7/e','rl7.5/e']
x3 = df4['district']
rrelist = ['rr1/e','rr2/e','rr3/e','rr4/e','rr5/e','rr6/e','rr7/e','rr7.5/e']

# 그래프 그리기
plt.figure(figsize=(15,18))

plt.subplot(3,2,1)
for i in rcelist:
  plt.plot(x1,df2[i])
  plt.scatter(x1,df2[i])
plt.xticks(rotation=90)
plt.ylabel('보수 투표율')
plt.title("서울 구별 '보수' 투표율")

plt.subplot(3,2,2)
for i in rcelist:
  plt.plot(x1,df2[i]-df2['c/e_ave'])
  plt.scatter(x1,df2[i]-df2['c/e_ave'])
plt.xticks(rotation=90)
plt.ylabel('보수 투표율 편차 - 구별 평균')
plt.title("서울 구별 '보수' 투표율의 편차")

plt.subplot(3,2,3)
for i in rlelist:
  plt.plot(x2,df3[i])
  plt.scatter(x2,df3[i])
plt.xticks(rotation=90)
plt.ylabel('진보 투표율')
plt.title("서울 구별 '진보' 투표율")

plt.subplot(3,2,4)
for i in rlelist:
  plt.plot(x2,df3[i]-df3['l/e_ave'])
  plt.scatter(x2,df3[i]-df3['l/e_ave'])
plt.xticks(rotation=90)
plt.ylabel('진보 투표율 편차 - 구별 평균')
plt.title("서울 구별 '진보' 투표율의 편차")
plt.legend(['1회','2회','3회','4회','5회','6회','7회','7.5회'],bbox_to_anchor=(1.08,0.8),fontsize=15)

plt.subplot(3,2,5)
for i in rrelist:
  plt.plot(x3,df4[i])
  plt.scatter(x3,df4[i])
plt.xticks(rotation=90)
plt.xlabel('구')
plt.ylabel('그 외 투표율')
plt.title("서울 구별 '그 외' 투표율")

plt.subplot(3,2,6)
for i in rrelist:
  plt.plot(x3,df4[i]-df4['r/e_ave'])
  plt.scatter(x3,df4[i]-df4['r/e_ave'])
plt.xticks(rotation=90)
plt.xlabel('구')
plt.ylabel('그 외 투표율 - 구별 평균')
plt.title("서울 구별 '그 외' 투표율의 편차")

plt.tight_layout()
plt.show()
```
